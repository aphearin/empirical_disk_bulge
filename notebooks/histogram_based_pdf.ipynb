```python
from empirical_disk_bulge.load_sdss_data import load_umachine_sdss_complete

sdss = load_umachine_sdss_complete()
mask = sdss['type_mendel13'] != 4
mask *= sdss['deltaBD_mendel13'] <= 1
cut_sdss = sdss[mask]
```

```python
from empirical_disk_bulge.observations.binned_pdfs import bt_pdf_fixed_sm_ssfr

npts_bt = 15
bt_abscissa = np.linspace(0, 1, npts_bt)
np.save('bt_abscissa', bt_abscissa)
bt_mids = 0.5*(bt_abscissa[:-1] + bt_abscissa[1:])
```

```python
fixed_sm, fixed_ssfr = 10, -10.
result1 = bt_pdf_fixed_sm_ssfr(cut_sdss['bt'], cut_sdss['sm'], cut_sdss['ssfr'], 
                     fixed_sm, fixed_ssfr, bt_abscissa,
                    sigma_sm=0.1, sigma_ssfr=0.25) 

fixed_sm, fixed_ssfr = 10, -10.75
result2 = bt_pdf_fixed_sm_ssfr(cut_sdss['bt'], cut_sdss['sm'], cut_sdss['ssfr'], 
                     fixed_sm, fixed_ssfr, bt_abscissa,
                    sigma_sm=0.1, sigma_ssfr=0.25) 

fixed_sm, fixed_ssfr = 10, -11.5
result3 = bt_pdf_fixed_sm_ssfr(cut_sdss['bt'], cut_sdss['sm'], cut_sdss['ssfr'], 
                     fixed_sm, fixed_ssfr, bt_abscissa,
                    sigma_sm=0.1, sigma_ssfr=0.25) 

fig, ax = plt.subplots(1, 1)

__=ax.plot(bt_mids, result1, color=blue)
__=ax.plot(bt_mids, result2, color=green)
__=ax.plot(bt_mids, result3, color=red)
```
